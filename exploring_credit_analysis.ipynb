### Dropando casos com ?

```python
indices_pra_remover = []

print(len(data))
for index, row in data.iterrows():
    if '?' in row.values:
        # print(row.values)
        # data.drop(row)
        indices_pra_remover.append(index)
data = data.drop(indices_pra_remover)
print(len(data))
```

```python
for column in data.columns:
    print(column, data[column].value_counts())
```

### Dropando outros casos alem do ?

```python
data = data.drop(data[data[3]=='l'].index)
data = data.drop(data[data[12]=='p'].index)
```

### Fazendo encoding

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

```

```python
for column in data.columns:
    if type(data[column].values[0]) == str:
        data[column] = le.fit_transform(data[column])
```

```python
data
```

```python
X = data.drop(columns = 15)
y = data[15]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf = RandomForestClassifier(max_depth=2, random_state=0)
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
y_test
```

```python
from sklearn.metrics import classification_report
target_names = ['class 0', 'class 1']
classification_report(y_test, y_pred, target_names=target_names)
```

```python
from sklearn import svm
```

```python
clf = svm.SVC()
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import classification_report
target_names = ['class 0', 'class 1']
classification_report(y_test, y_pred, target_names=target_names)
```
